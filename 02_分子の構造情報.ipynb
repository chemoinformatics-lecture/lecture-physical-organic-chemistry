# 分子の読み込み

```python
import openbabel
from openbabel import pybel
```

```python
m = pybel.readstring("smiles", "CC(C)(C)C1CCCCC1")
```

```python
m
```

```python
print(m.write("xyz"))
```

```python
m.make3D(forcefield='mmff94', steps=1000)
print(m.write("sdf"))
```

```python
print(m.write("xyz"))
```

```python
import py3Dmol
```

```python
# stickとボール
view = py3Dmol.view()
view.addModel(m.write("xyz"),'xyz')
view.setStyle({'line':{}})
view.zoomTo()
view.show()
```

```python
# stickとボール
view = py3Dmol.view()
view.addModel(m.write("xyz"),'xyz')
view.setStyle({'stick':{}})
view.zoomTo()
view.show()
```

```python
view = py3Dmol.view()
view.addModel(m.write("xyz"),'xyz')
view.setStyle({'sphere':{}})
view.zoomTo()
view.show()
```
